## ML with Load_Boston()
#### 1. Import python packages

```python
import numpy as np 
import pandas as pd 
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
```

#### 2. Load data Boston

```python
dataBoston = load_boston()

df = pd.DataFrame(
    dataBoston['data'],
    columns = dataBoston['feature_names']
)
df['PRICE'] = dataBoston['target']
df
```

```python
corr = df.corr()
ticks = np.arange(12)
plt.imshow(corr, cmap = 'BuPu_r')
plt.colorbar()
plt.set_xticks(ticks, df.columns)
plt.show()
```
